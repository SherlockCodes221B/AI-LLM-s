```python
import pandas as pd
import neattext as nf
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/AI/Course_recommendation/udemy_course_data.csv')
```

```python
data.head()
```

```python
data['course_title_cleaned'] = data['course_title'].apply(nf.remove_stopwords)
data['course_title_cleaned'] = data['course_title'].apply(nf.remove_special_characters)
data.head()
```

```python
cf = CountVectorizer()
count_vec = cf.fit_transform(data['course_title_cleaned'])
count_vec.todense()
```

```python
cv_words = pd.DataFrame(count_vec.todense(),columns = cf.get_feature_names_out())
cv_words
```

```python
cos = cosine_similarity(cv_words)
```

```python
cos
```

```python
course_index = pd.Series(data.index,index = data['course_title']).drop_duplicates()
```

```python
course_index
```

```python
data.columns
```

```python
scores = list(enumerate(cos[course_index['How To Maximize Your Profits Trading Options']]))
scores
```

```python
sorted_score = sorted(scores, key = lambda x: x[1], reverse = True)
sorted_score
```

```python
sorted_indices = [i[0] for i in sorted_score[1:]]

sorted_values = [i[1] for i in sorted_score[1:]]
```

```python
recommend_df = data.iloc[sorted_indices]
```

```python
import numpy as np
recommend_df['Similarity_Score'] = np.array(sorted_values)
recommend_df
```

```python
usable = recommend_df[['course_title_cleaned','Similarity_Score']]
usable
```

```python
def recommend_a_course(title,number_of_record=10):
  course_index = pd.Series(data.index, index=data['course_title']).drop_duplicates()
  index = course_index[title]
  scores = list(enumerate(cos[index]))
  sorted_scores = sorted(scores,key = lambda x:x[1],reverse=True)
  selected_course_index = [i[0] for i in sorted_scores[1:]]
  selected_course_score = [i[1] for i in sorted_scores[1:]]
  rec_df = data.iloc[selected_course_index]
  rec_df['Similarity_Score'] = selected_course_score
  final_recommendations = rec_df[[ 'course_title', 'Similarity_Score', 'url', 'price', 'num_subscribers']]
  return final_recommendations.head(number_of_record)

```
